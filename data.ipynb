```python
import pandas as pd

path = "/Users/realdragon/Awesome J Dropbox/Master Jibetcenter/Mac (3)/data/"
course_simple = pd.read_csv(path+"course.csv")
course_detail = pd.read_csv(path + "course_data.csv", skiprows=[57908])

course_detail.head()
```

```python
drop = ["id", "num_reviews_recent", "published_time", "_class", "is_paid", "price_detail", "image_125_H", "image_240x135", "image_480x270", "published_title", "num_subscribers", "avg_rating_recent", "content_info", "prerequisites", "objectives", "rating_distribution", "num_published_lectures", "num_published_quizzes", "num_of_published_curriculum_objects", "status_label"]
```

```python
course_detail.drop(drop, axis=1, inplace=True)
course_detail.head()
```

```python
course_detail["content_info_short"] = course_detail["content_info_short"].str.split(" ").str[0]
course_detail.head()
```

```python
# library
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

X = course_detail[['avg_rating', 'num_reviews', 'content_info_short']].values
km = KMeans(n_clusters=10, random_state=21)
km.fit(X)
```

```python
plt.scatter(X[:,0], X[:,1], c=km.labels_, cmap='gist_rainbow')
plt.xlabel('Average Rating')
plt.ylabel('Number of Reviews')
```

```python
course_detail = course_detail.dropna(subset=['avg_rating', 'num_reviews', 'content_info_short'])
course_detail.tail()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X = course_detail[['avg_rating', 'num_reviews', 'content_info_short']].values

wcss = []
for i in range(1, 11):  # 클러스터 개수 1~10까지 시도
    km = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10)
    km.fit(X)
    wcss.append(km.inertia_)  # WCSS 값 저장

# Elbow method graph
plt.figure(figsize=(10,5))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') 
plt.show()

km = KMeans(n_clusters=5, random_state=21)
km.fit(X)

# scatter plot
plt.scatter(X[:,0], X[:,1], c=km.labels_, cmap='gist_rainbow')
plt.xlabel('Average Rating')
plt.ylabel('Number of Reviews')
```
